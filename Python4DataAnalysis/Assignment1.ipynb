# Question 1.1:
Create a list of numbers which are multiple of 3 and
7 between 1 to 500 (5 points).


```python
s= list(range(1,501))
mylist=[]
for num in s:
    if num % 3 == 0 and  num % 7 == 0 :
                  
        mylist.append(num)
            
print("List of numbers divisible are :" + str(mylist))
```

# Question 1.2:
For the list, find what is the mean. Print what is 25th
quartile, 50th quartile and 75th quartile. (10 points)


```python
s= list(range(1,501))
mylist=[]
for num in s:
    if num % 3 == 0 and  num % 7 == 0 :
                  
        mylist.append(num)
            
print(mylist)
print("Number of numbers : " + str(len(mylist)))
mean = float(sum(mylist)/max(len(mylist),1))
print ("Mean           : " + str(mean))

index= (len(mylist)) // 2

firstQuartile=(len(mylist)) // 4
print("firstQuartile  : " + str(mylist[firstQuartile]))

if (len(mylist) % 2):
    print("secondQuartile : " + str(mylist[index]))
else:
    print("secondQuartile : " + str((mylist[index] + mylist[index + 1])/ 2.0))
    
thirdQuartile=(3 * len(mylist)) // 4
print("thirdQuartile  : " + str(mylist[thirdQuartile]))
```

# Question 2.2:

Find out what is the average, median and mode for
word length (character length) (10 points).


```python
file=open("/Users/Rohits216/Desktop/DataAnalysisPython-master/Lecture-03/ques2.txt")
count=0
l=0
S=[0]
corresponding={}
occurances=[]

for line in file:    
    for word in line.split():
         S.append(len(word))
         S.sort()
         l += len(word)
         count+=1 
 
    #print(S)        
print("Total word length: " + str(l))
print("Average word length: " + str(l/count))
index= (len(S)) // 2
if (len(S) % 2):
    print("Median : " + str(S[index]))
else:
    print("Median : " + str((S[index] + S[index + 1])/ 2.0))
    
for i in S:
        count = S.count(i)
        corresponding.update({i:count})

for i in corresponding:
        freq=corresponding[i]
        occurances.append(freq)


values=corresponding.values()
index= (len(corresponding))
#print(values)

print("Mode   : " + str(len(str(max(values)))))
```
